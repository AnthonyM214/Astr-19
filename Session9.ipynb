> **Create a Jupyter notebook, import matplotlib. Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.**



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
RandNums = np.random.uniform(0, 1, 1000) #pull 1000 random uniform numbers[0..1]

plt.hist(RandNums, bins =100, color = "yellowgreen", edgecolor="white")

  #label my axes
plt.ylabel('Y label')
plt.xlabel('Slime Concentration')

plt.show()
plt.savefig("Session9.pdf")
```
