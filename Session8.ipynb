```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100) #Array ranging from [0,1] in 100 steps

def sinX(x):
    return np.sin(x)

def cosX(x):
    return np.cos(x) # Defined Sin and Cos functions..


```

```python
plt.figure(figsize=(8,4))
plt.subplot(1, 2, 1) #straight from matplotlib (1 Row, 2 Columns)
plt.plot(x,sinX(x),color="navy")
plt.xlabel("x")
plt.ylabel("Sin(x)")

plt.subplot(1, 2, 2) #2nd Column #last conditional creates a plot alongside
plt.plot(x,cosX(x),color="tomato")
plt.xlabel("x")
plt.ylabel("Cos(x)")
```
